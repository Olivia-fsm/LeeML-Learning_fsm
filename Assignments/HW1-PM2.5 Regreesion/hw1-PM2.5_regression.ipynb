## Extract Feature(2)
**Pseudo Code**
- Declare train_X for previous 9-hr data, and train_y for <u>10th-hr pm2.5</u> value
- For i in all the given data:
    sample every 10-hrs:
        train_x.append(previous 9-hr data)
        train_y.append(the value of 10th-hr pm2.5)
  add a bias term to every data in train_x

#### Explanation:
每个月有480hr的数据， 每9hr生成一个training data，因此每个月有471组training data，故总资料数为471\*12组；每组有（9hr* 18elements）个feature。

对应有（471* 12）组target，为第十个小时的PM2.5的值。

```python
x = np.empty([12*471, 18*9], dtype = float)
y = np.empty([12*471, 1], dtype = float)
for month in range(12):
    for countSet in range(471):
        for countHR in range(9):
            x[month*471+countSet, 18*countHR:18*(countHR+1)] = monthData[f'month{month+1}'][:, countSet].T
            y[month*471+countSet, 0] = monthData[f'month{month+1}'][:,countSet+9][9]
    
    
```

## Normalize(1)


```python
meanX = np.mean(x, axis = 0)
stdX = np.std(x, axis = 0)
for i in range(len(x)):
    for j in range(len(x[0])):
        if(stdX[j] != 0):
            x[i][j] = (x[i][j]-meanX[j])/(stdX[j])
```

## Implement Linear Regression!

**Pseudo Code**
1. Declare weight vector, initial learning_rate(lr), and # of iteration
2. for i_th iteration:

    y'=the inner product of train_x and weight vector;
    
    Loss = y'-y_train;
    
    gradient = 2* np.dot((x_train)', L);
    
    weight vector -= learning rate(lr)* gradient.

```python
print(x.shape) # [12*471, 18*9]
print(x[0])
```

```python
def linearRegression(x_train_orig, y_train,x_test_orig, y_test, epoch=2500, lr=0.00000015):
    dims = x_train_orig.shape[1]+1
    weights = np.random.randn(dims, 1)
    x_train = np.concatenate((np.ones([x_train_orig.shape[0], 1]), x_train_orig), axis = 1).astype(float)
    x_test = np.concatenate((np.ones([x_test_orig.shape[0], 1]), x_test_orig), axis = 1).astype(float)
    Costs = []
    Losses = []
    weightsList = []
    for epoch in range(1,epoch+1):
        weightsList.append(weights)
        y_hat = np.dot(x_train, weights)
        cost = np.sum(np.power((y_hat - y_train), 2))/(x_train.shape[0])
        Costs.append(cost)
        y_pred = np.dot(x_test, weights)
        loss = np.sum(y_pred - y_test)/(x_test.shape[0])
        Losses.append(loss)
        grad = 2*np.dot(x_train.T,(y_hat - y_train))
        weights = weights-lr*grad
        if (epoch%100 == 0):
            print(f"The Loss of {epoch}'s iteration: {loss}")
    parameters = {}
    parameters['Costs'] = Costs
    parameters['Losses'] = Losses
    parameters['weights'] = weightsList
        
    return parameters
        
    
    
    
```

```python
parameters = linearRegression(x_train, y_train,x_test,y_test)
```

## Adagrad

**Pseudo Code**
1. Declare weight vector, initial lr, and # of iteration
    **Declare prev_gra storing gradients in every previous iterations!**
2. For i_th iteration:

    y'=the inner product of x_train and weight vector 
    
    Loss = y'-y_train
    
    gradient = 2* np.dot((x_train)', L)
    
    **prev_gra  += gra ** 2 **
    
    **ada = np.sprt(prev_gra)**
    
    weight vector -= learning rate(lr)* gradient **/ada**
    
    

```python
def linearRegression_Adagrad(x_train_orig, y_train,x_test_orig, y_test, epoch=25000, lr=100, eps = 0.0000001):
    dims = x_train_orig.shape[1]+1
    weights = np.random.randn(dims, 1)
#     weights = np.zeros([dims, 1])
    x_train = np.concatenate((np.ones([x_train_orig.shape[0], 1]), x_train_orig), axis = 1).astype(float)
    x_test = np.concatenate((np.ones([x_test_orig.shape[0], 1]), x_test_orig), axis = 1).astype(float)
    Costs = []
    Losses = []
    weightsList = []
    grads = []
    for epoch in range(1,epoch+1):
        weightsList.append(weights)
        y_hat = np.dot(x_train, weights)
        cost = np.sum(np.power((y_hat - y_train), 2))/(x_train.shape[0])
        Costs.append(cost)
        y_pred = np.dot(x_test, weights)
        loss = np.sum(y_pred - y_test)/(x_test.shape[0])
        Losses.append(loss)
        grad = 2*np.dot(x_train.T,(y_hat - y_train))
        grads.append(grad)
        vt = np.sum(np.power(grads, 2))
        weights = weights-lr*grad/(np.sqrt(vt+eps))
        if (epoch%100 == 0):
            print(f"The Loss of {epoch}'s iteration: {cost}")
    parameters = {}
    parameters['Costs'] = Costs
    parameters['Losses'] = Losses
    parameters['weights'] = weightsList
        
    return parameters
        
    
    
    
```

```python
parameters = linearRegression_Adagrad(x_train, y_train,x_test,y_test)
```
